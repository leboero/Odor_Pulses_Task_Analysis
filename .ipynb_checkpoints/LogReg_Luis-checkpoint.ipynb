```python
def validator(trials,n_reps,n_bins, delay): # x_array (odor profile passed through the kernel) 
        #prediction on z_array (licking decision)
        
    logreg = LogisticRegression(max_iter=400)
    
    #Creating empty arrays and list where the results of the regression will be appended
    
    coef = np.zeros((n_reps,n_bins+3))
    icept = np.zeros((n_reps,n_bins+3))
    acc_score = []
    prec_score = []
    rec_score = []
    specificity = []
    p_res = []
    predict_proba = []
    
    lr_f1 = np.zeros(n_reps)
    lr_auc = np.zeros(n_reps)
    auc_roc = np.zeros(n_reps)

    logreg_loss = np.zeros(n_reps)
    
    #For pd.DataFrame creation below. Accessory
    
    colx=[]
    for i in range(n_bins):
        colx.append('Bin_'+str(i))
        
    last_cols = ['animal_id', 'session_n', 'trial_n']
    #colx.extend(last_cols)
    
    #Empty arrays and lists for managing predicting variables
    
    nsamples = len(trials)
    lick_array = np.zeros(nsamples)
    x_raw = []
    trial_n = []
    session_n = []
    animal_identity = []
    
    for k in range(nsamples): #Getting licking (y) and all predictor vairables information
        #(X) from each trial
        
        lick_array[k] = trials[k].lick_side #(predicted variable, y)
        splitted_x = np.array_split(get_odor_profile_actual(trials[k])[0:delay], n_bins)
        #splitting the odor profile in bins following the n_bins input
        trial_n.append(trials[k].number)
        session_n.append(trials[k].session)
        animal_identity.append(trials[k].name)
        
        for j in range(len(splitted_x)):
            x_raw.append(splitted_x[j].mean()) #Getting the mean (for dimensionality reduction)
            # of the odor profile in each bin and passing that to a list
        
    #Converting all the predictor variable lists to numpy arrays
    
    x_raw = np.asarray(x_raw, dtype="object")
    trial_n = np.asarray(trial_n, dtype="object")
    session_n = np.asarray(session_n, dtype="object")
    animal_identity = np.asarray(animal_identity, dtype="object")
    a_id = np.unique(animal_identity, return_inverse=True)[1]
    
    #Reshaping all the predictor variable numpy arrays and appending them to make an X array

    x_raw = x_raw.reshape((nsamples,n_bins))
    trial_n = trial_n.reshape((nsamples,-1))
    session_n = session_n.reshape((nsamples,-1))
    animal_id = a_id.reshape((nsamples,-1))
    
    append_1 = np.append(x_raw, animal_id, axis=1)
    append_2 = np.append(append_1, session_n, axis=1)
    final_append = np.append(append_2, trial_n, axis=1)
    
    for i in tqdm.notebook.trange(n_reps):
        
        #Initial shuffle of trial order before splitting into train/testHi,
        
        ind_array = np.random.permutation(nsamples)
        x_shuffled = final_append[ind_array,:]
        z_shuffled = lick_array[ind_array].squeeze()
        
        #Train,test,split + fitting the logistic regression
        
        X_train, X_test, y_train, y_test = train_test_split(x_shuffled, z_shuffled, test_size=0.2, random_state=42)
        logreg.fit(X_train,y_train)
        
        #Getting coefficients and intercept
        
        co = logreg.coef_[0]
        ipt = logreg.intercept_[0]
        
        #Getting Pearson Residuals
        
        pred_values = logreg.predict(X_test)
        pp = logreg.predict_proba(X_test)
        #print(pp.shape, pp)
        y_pred_proba = pp[:,1]
        logreg_loss[i] = log_loss(y_test, pp)
        #print(logreg_loss)
        pearson_residuals = (y_test - y_pred_proba) / np.sqrt(y_pred_proba * (1-y_pred_proba))
        predict_proba.append(y_pred_proba)
        p_res.append(pearson_residuals)
        
        #Generating ROC and Precision-Recall curves
        lr_precision, lr_recall, _ = sklearn.metrics.precision_recall_curve(y_test, y_pred_proba)
        lr_f1[i], lr_auc[i]  = sklearn.metrics.f1_score(y_test, pred_values), sklearn.metrics.auc(lr_recall, lr_precision)
        auc_roc[i] = sklearn.metrics.roc_auc_score(y_test, y_pred_proba)
        
#         #Calculating Hessian Matrix and p_values of coefficients
#         X_train_bias = np.column_stack((np.ones(X_train.shape[0]), X_train))
#         p = logreg.predict_proba(X_train)
#         W = np.diagflat(p[:, 1]*(1-p[:,1]))
#         hessian = np.linalg.inv(X_train_bias.T @ W @ X_train_bias)
#         se = np.sqrt(np.diag(hessian))
#         z_scores = coef/se
#         p_values = 2*(1-norm.cdf(np.abs(z_scores)))
#         print(p_values.shape)
        
        #Getting accuracy of the model
        acc = sklearn.metrics.accuracy_score(y_test, pred_values)
        preci = sklearn.metrics.precision_score(y_test, pred_values)
        reca = sklearn.metrics.recall_score(y_test, pred_values)
        acc_score.append(acc)
        prec_score.append(preci)
        rec_score.append(reca)
        
        bin_w = logreg.coef_[0][:n_bins]
        other_w = logreg.coef_[0][n_bins:]
        sum_w = logreg.coef_[0][:n_bins].sum()
        rel_w = bin_w/sum_w
        coef[i] = np.concatenate((rel_w, other_w), axis=0).squeeze()
        #coef[i] = logreg.coef_[0]
        icept[i] = logreg.intercept_[0]
        
    coef_df = pd.DataFrame(coef[:,0:n_bins], columns=colx)

    df = pd.melt(coef_df, var_name='Bin', value_name='Rel_w')
    coef_t = np.transpose(coef)
    
    pearson_res = np.asarray(p_res, dtype=np.float64)
    y_predict = np.asarray(predict_proba, dtype=np.float64)
    acc_score = np.asarray(acc_score, dtype="object").mean()
    precision_score = np.asarray(prec_score, dtype="object").mean()
    recall_score = np.asarray(rec_score, dtype="object").mean()
    pr_auc = lr_auc.mean()
    auc_roc_curve = auc_roc.mean()
    auc_roc_curve_std = auc_roc.std()
    
    ci_weights = np.zeros((coef.shape[1],2))
    lower_bound = np.zeros(coef.shape[1])
    upper_bound = np.zeros(coef.shape[1])
    final_coef = coef.mean(axis=0)
    final_coef_std = coef.std(axis=0)
    final_pearson_res = pearson_res.mean(axis=0)
    final_pearson_res_sem = sem(pearson_res, axis=0)

    for q in range(coef.shape[1]):
        
        ci = pg.compute_bootci(coef_t[q], func='mean', n_boot=1000, decimals=3)
        lower_bound[q] = final_coef[q] - ci[0]
        upper_bound[q] = ci[1] - final_coef[q]
        
    actual_ci = np.hstack((lower_bound, upper_bound))
    final_ci = actual_ci.reshape(coef.shape[1],2, order='F')
    
    loss_mean = logreg_loss.mean()
    loss_std = logreg_loss.std()
    
    #plt.scatter(y_predict, pearson_res)
    #print(colx)
    return final_coef, final_coef_std, df, acc_score, auc_roc_curve, auc_roc_curve_std, loss_mean, loss_std
```

```python
def get_by_animal(trial_list, name):
    return list(filter(lambda trial: trial.name==name,trial_list))

def get_sessions(trial_list):
    sessions = list()
    for trial in trial_list:
        if not trial.session in sessions:
            sessions.append(trial.session)
    sessions.sort()
    return sessions

def get_by_session(trial_list, session):
    return list(filter(lambda trial: trial.session==session,trial_list))

def get_by_date(trial_list, date):
    date = str(date)
    return list(filter(lambda trial: trial.date==date,trial_list))

def get_by_session_type(trial_list, session_type):
    return list(filter(lambda trial: trial.session_type==session_type,trial_list))

def get_performance(trial_list):
    total = 0.0
    for trial in trial_list:
        total += int(trial.correct)
    return total/len(trial_list)

def get_correct_trials(trial_list):
    total = 0.0
    for trial in trial_list:
        if trial.correct == True:
            print(1)
            
        else:
            print(0)

def get_by_delay(trial_list, delay):
    return list(filter(lambda trial: trial.delay_time==delay,trial_list))

def get_by_contrast(trial_list, contrast):
    return list(filter(lambda trial: round(trial.high_count/trial.low_count)==contrast,trial_list))

def get_performance_list(trials):
    sessions = get_sessions(trials)
    performance_list = list()
    for session in sessions:
        performance_list.append(get_performance(get_by_session(trials,session)))
    return performance_list

def butter_lowpass(cutoff, fs, order=5):
    nyq = 0.5 * fs
    normal_cutoff = cutoff / nyq
    b, a = butter(order, normal_cutoff, btype='low', analog=False)
    return b, a

def butter_lowpass_filter(data, cutoff, fs, order=5):
    b, a = butter_lowpass(cutoff, fs, order=order)
    y = filtfilt(b, a, data)
    return y

def get_inhalation(trial):
    breath = butter_lowpass_filter(trial.breath_profile, cutoff = 10.0, fs = 1000.0, order = 6)
    breath = breath - breath.mean()
    breath = breath / (breath.max() - breath.min()) * 2
    inhalation = np.gradient(breath) < -0.00
    return inhalation.astype(float)

def get_odor_profile_actual(trial):
    kernel = np.load('odor_kernel_50ms.npy')
    pulse_onsets = (np.diff(trial.odor_profile)>0).astype(float)
    pulse_onsets_shifted = np.concatenate((np.zeros((70,)),pulse_onsets,np.zeros(131)))
    odor_profile_actual = np.convolve(pulse_onsets_shifted,kernel,mode='same')
    return odor_profile_actual

def get_full_inhalation(trial):
    breath = butter_lowpass_filter(np.concatenate((trial.pre_breath_profile,trial.breath_profile)), 
                                   cutoff = 10.0, fs = 1000.0, order = 6)
    breath = breath - breath.mean()
    breath = breath / (breath.max() - breath.min()) * 2
    inhalation = np.gradient(breath) < -0.00
    return inhalation.astype(float)

def get_full_inhalation_test(trial):
    breath = butter_lowpass_filter(np.concatenate((trial.pre_breath_profile,trial.breath_profile)), 
                                   cutoff = 10.0, fs = 1000.0, order = 6)
    breath = breath - breath.mean()
    breath = breath / (breath.max() - breath.min()) * 2
    breath[breath>0] = 0
    return -breath
    

def get_inhaled_odor(trial):
    inhalation = get_inhalation(trial)
    odor_profile = get_odor_profile_actual(trial)
    total_odor = (inhalation[0:odor_profile.size]*odor_profile).sum()
    return total_odor

def get_inhaled_odor_trace(trial):
    sniff_mod_kernel = np.load('inhalation_kernel_fine_weights_active.npy')

    delay_time = trial.delay_time
    sniff = butter_lowpass_filter(trial.breath_profile,10,1000,3)[0:delay_time]
    sniff = (sniff - sniff.mean() +1)/2
    sniff_onset,_ = find_peaks(sniff,distance=150)
    num_sniffs = len(sniff_onset)
    odor_profile_actual = get_odor_profile_actual(trial)
    sniff_norm = np.zeros((num_sniffs,250))
    odor_norm = np.zeros((num_sniffs,250))

    sniff_marks = np.concatenate((sniff_onset[:-1].reshape(-1,1),sniff_onset[1:].reshape(-1,1)),axis=1)
    sniff_durations = np.diff(sniff_marks,axis=1)
    num_sniffs = sniff_marks.shape[0]
    sniff_mod = np.zeros(sniff.shape)
    for i in range(num_sniffs):
        sniff_mod[int(sniff_marks[i,0]):int(sniff_marks[i,1])] = resample(sniff_mod_kernel,int(sniff_durations[i]))
        
    inhaled_odor_trace =  odor_profile_actual[0:delay_time]*sniff_mod
    return inhaled_odor_trace

def get_by_pulses(trial_list, low_pulses, high_pulses):
    return list(filter(lambda trial: (trial.num_pulses>=low_pulses and trial.num_pulses<high_pulses),
                       trial_list))

def get_by_inhaled(trial_list, low, high):
    return list(filter(lambda trial: (get_inhaled_odor(trial)>=low and get_inhaled_odor(trial)<high),
                       trial_list))

def sigmoid (x):
    return 1/(1+np.exp(-x))

def z_stat(nbins, w1, w2, w1sd, w2sd):
    b = w1[0:nbins] - w2[0:nbins]
    se1 = w1sd[0:nbins]**2
    se2 = w2sd[0:nbins]**2
    sc = np.sqrt(se1+se2)
    z = b / sc
    p_values = scipy.stats.norm.sf(abs(z))*2
    corrected_p_values = smt.multipletests(p_values, alpha=0.05, method='sidak')
    return corrected_p_values[0], corrected_p_values[1] 
```
